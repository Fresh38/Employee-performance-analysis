```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# Identify the categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
print("\nCategorical columns:")
print(categorical_columns)
```

```python
# To Initialize LabelEncoder
label_encoders = {}
```

```python
# Applying Label Encoding to all categorical columns
for column in categorical_columns:
    if df[column].nunique() < 10:  
        label_encoders[column] = LabelEncoder()
        df[column + '_Label'] = label_encoders[column].fit_transform(df[column])
    else:
        # For columns with more than 10 unique values, using One-Hot Encoding
        one_hot_encoded = pd.get_dummies(df[column], prefix=column)
        df = pd.concat([df, one_hot_encoded], axis=1)
        df.drop(column, axis=1, inplace=True)

print("\nDataFrame after encoding:")
print(df.head())
```

```python
# Computing the correlation matrix
correlation_matrix = df.corr()
```

```python
# Preview the correlation matrix
print("\nCorrelation matrix:")
print(correlation_matrix)
```

```python
# Determining the variable most highly correlated with the performance rating

performance_corr = correlation_matrix['PerformanceRating'].sort_values(ascending=False)
print("\nCorrelation with PerformanceRating:")
print(performance_corr)
```

```python
# Any other significant relationships
significant_corr = correlation_matrix[(correlation_matrix > 0.5) | (correlation_matrix < -0.5)]
print("\nSignificant correlations (absolute value > 0.5):")
print(significant_corr)
```
